```python
sns.histplot(data=df, x="Age", hue="Survived", bins=30, palette={0:"red", 1:"green"})
plt.show()
```

```python
#Create age bins
bins = [0, 12, 18, 35, 50, 80]  # define ranges
labels = ['Child', 'Teen', 'Adult', 'Middle Age', 'Senior']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Check the new column
df[['Age','AgeGroup']].head()
```

```python
age_survival = pd.crosstab(df['AgeGroup'], df['Survived'])
print(age_survival)
```

```python
age_survival.plot(kind='bar', stacked=True, color=['orange','blue'])
plt.title("Survival by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Passengers")
plt.show()

#orange is more deaths, blue is survived
```

```python
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

- Passengers in 1st class had the highest survival rate.
- Those in 3rd class had the lowest survival rate.
- This highlights the impact of socio‑economic status on survival chances.


```python
sns.histplot(data=df, x="Fare", hue="Survived", bins=30, palette={0:"red", 1:"green"})
plt.title("Fare Distribution by Survival")
plt.show()
```

- Passengers who paid higher fares (often 1st class) had better survival rates.
- Lower‑fare passengers (mostly 3rd class) had lower survival chances.
- This reinforces the socio‑economic impact on survival.


```python
sns.countplot(x="Embarked", hue="Survived", data=df)
plt.title("Survival by Embarked Port")
plt.show()
```

- Passengers who embarked from Cherbourg (C) generally had higher survival rates.
- Those from Southampton (S) had lower survival chances.
- This suggests embarkation location may be linked to socio‑economic status or cabin allocation.




```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=['int64','float64']).corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

- Survived vs Fare → Positive correlation
- Passengers who paid higher fares (often 1st class) had better survival chances.
- Survived vs Pclass → Negative correlation
- Higher passenger class number (3rd class) is linked with lower survival.
- In other words, 1st class passengers were more likely to survive.
- Survived vs Age → Very weak correlation
- Age alone doesn’t strongly predict survival, but grouping into categories (children vs adults) reveals clearer patterns.
- SibSp (siblings/spouses aboard) & Parch (parents/children aboard)
- Small correlations with survival. Having family aboard sometimes increased survival chances, but not strongly.
- Pclass vs Fare → Strong negative correlation
- Higher class (1st) is associated with higher fares, while 3rd class tickets were cheaper.
